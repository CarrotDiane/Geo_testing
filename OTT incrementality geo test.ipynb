## Simulate on an out-of-sample period

```python
# define pre-test period and simulation period (as out-of-sample to simulate a test period)

import datetime
pre_test_start = datetime.datetime.strptime('2023-01-01', '%Y-%m-%d').date()
pre_test_end = datetime.datetime.strptime('2023-02-01', '%Y-%m-%d').date()

simulate_test_start = datetime.datetime.strptime('2023-02-02', '%Y-%m-%d').date()
simulate_test_end = datetime.datetime.strptime('2023-02-12', '%Y-%m-%d').date()

print('Number of days in out-of-sample period:', (simulate_test_end - simulate_test_start).days + 1)
```

```python
print("Expected Lift:", r.minimum_detectable_effect)

r.simulate(data = data, best_candidate = res['candidate_0'], 
           test_group = 'group0', control_group = 'group1', 
           metric = 'dma_agg_mp_imao_12m', 
           pre_test_start = pre_test_start, pre_test_end = pre_test_end, 
           simulate_test_start = simulate_test_start, simulate_test_end = simulate_test_end, 
           lift = r.minimum_detectable_effect, 
           alpha = 0.05, power = 0.8)
```
